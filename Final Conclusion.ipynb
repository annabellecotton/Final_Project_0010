**Primary Care Access Data**

```python
import pandas as pd

# load in the primary care file
data = pd.read_csv("primary_care_access.csv", index_col="PRACTICE_ZIP")
#value count
full_count = data['PRACTICE_CITY'].value_counts() 
#drop Pittsburgh and tally neighborhoods
count = full_count.drop("Pittsburgh")
print(count)
##set size of plots
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 5]
##plot the bar graph
count.plot.bar()
```

**Air Quality Data**

```python
air_quality_pandas = pd.read_csv("air_quality.csv", index_col="bp")
count = air_quality_pandas['site'].value_counts() 
print(count)
# set size of plots
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']

# plot the bar graph
count.plot.bar()
```

**Combining the Data**

To combine the data, we assigned the discovered values to a graph and created a ranking system. The rank was based on the highest value in the data set. In the Primary Care access, the highest value was 30, so that was assigned to a ranking of "1". After that, the rest of the rankings were based out of that value (neighborhood primary care amount / 30 = ranking). The air quality rankings were based in a similiar way. The lowest value in the data set is the best air quality. So the lowest value (8694) was assigned as a "1" and the rest of the  rankings were based out of this number (8694 / air quality = air quality ranking).
